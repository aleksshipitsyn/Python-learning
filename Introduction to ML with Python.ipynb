# Introduction to ML with Python

## Working with Text Data

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_files

reviews_train = load_files("data/aclImdb/train/")
```

```python
text_train, y_train = reviews_train.data[:25000], reviews_train.target[:25000]
print('Type of text_train: {}'.format(type(text_train)))
print('Length of text_train: {}'.format(len(text_train)))
print('Text_train[1]:\n{}'.format(text_train[1]))
```

```python
text_train = [doc.replace(b"<br />", b" ") for doc in text_train]
```

```python
print("Samples per class (training): {}".format(np.bincount(y_train)))
```

```python
reviews_test = load_files("data/aclImdb/test/")
```

```python
text_test, y_test = reviews_test.data[:25000], reviews_test.target[:25000]

print('Number of documents in test data: {}'.format(len(text_test)))
print('Samples per class (test): {}'.format(np.bincount(y_test)))

text_test = [doc.replace(b"<br />", b" ") for doc in text_test]
```

```python
bards_words = ['The fool doth think he is wise,', 
               'but the wise man knows himself to be a fool']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
vect.fit(bards_words)
```

```python
print("Vocabulary size: {}".format(len(vect.vocabulary_)))
print("Vacabulary content:\n{}".format(vect.vocabulary_))
```

```python
bag_of_words = vect.transform(bards_words)
print('bag of words: {}'.format(repr(bag_of_words)))
```

```python
print("Dense representation of bag_of_words:\n{}".format(bag_of_words.toarray()))
```

```python
vect = CountVectorizer().fit(text_train)
X_train = vect.transform(text_train)
print('X_train:\n{}'.format(repr(X_train)))
```

```python
feature_names = vect.get_feature_names()
print('Number of features: {}\n'.format(len(feature_names)))
print('First 20 features:\n{}\n'.format(feature_names[:20]))
print('Features 20010 to 20030:\n{}\n'.format(feature_names[20010:20030]))
print('Every 2000th feature:\n{}'.format(feature_names[::2000]))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

scores = cross_val_score(LogisticRegression(), X_train, y_train, cv=3)
print('Mean cross-validation accuracy: {:.2f}'.format(np.mean(scores)))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 1, 10]}
grid = GridSearchCV(LogisticRegression(), param_grid, cv=3)
grid.fit(X_train, y_train)

print('Best cv-score: {:.2f}'.format(grid.best_score_))
print('Best parameters: ', grid.best_params_)
```

```python
X_test = vect.transform(text_test)
print('{:.2f}'.format(grid.score(X_test, y_test)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline

pipe = make_pipeline(TfidfVectorizer(min_df=5, norm=None), LogisticRegression())

param_grid = {'logisticregression__C': [0.001, 0.01, 0.1, 10]}
grid = GridSearchCV(pipe, param_grid, cv=3)
grid.fit(text_train, y_train)

print('Best cross-validation score: {:.2f}'.format(grid.best_score_))
```

```python
vectorizer = grid.best_estimator_.named_steps['tfidfvectorizer']
X_train = vectorizer.transform(text_train)
```

```python
# find max for each feature in dataset
max_value = X_train.max(axis=0).toarray().ravel()
sorted_by_tfidf = max_value.argsort()
feature_names = np.array(vectorizer.get_feature_names())

print('Features with lowest tfidf:\n{}'.format(feature_names[sorted_by_tfidf[:20]]))
print('\nFeatures with highest tfidf:\n{}'.format(feature_names[sorted_by_tfidf[-20:]]))

```

```python
sorted_by_idf = np.argsort(vectorizer.idf_)
print('Features with lowest idf:\n{}'.format(feature_names[sorted_by_idf[:100]]))
```

```python
import mglearn
mglearn.tools.visualize_coefficients(
    grid.best_estimator_.named_steps['logisticregression'].coef_[0,],
    feature_names, n_top_features=40)
```

```python
import spacy, nltk

```

```python
# spacy's English-language models
en_nlp = spacy.load('en_core_web_sm')

# nltk Porter stemmer
stemmer = nltk.stem.PorterStemmer()
```

```python
def compare_normalization(doc):
    # tokenize with spacy
    doc_spacy = en_nlp(doc)
    print('Lemmatization:\n')
    print([token.lemma_ for token in doc_spacy])
    
    # tokenize with Porter
    print('\nStemming:\n')
    print([stemmer.stem(token.norm_.lower()) for token in doc_spacy])
    
```

```python
compare_normalization("Our meeting today was worse than yesterday, "
                      "I'm scared of meeting the clients tomorrow.")
```

```python
import re

regexp = re.compile('(?u)\\b\\w\\w+\\b')

en_nlp = spacy.load('en_core_web_sm')
old_tokenizer = en_nlp.tokenizer
en_nlp.tokenizer = lambda string: old_tokenizer.tokens_from_list(regexp.findall(string))

def custom_tokenizer(document):
    doc_spacy = en_nlp(document)
    return [token.lemma_ for token in doc_spacy]

lemma_vect = CountVectorizer(tokenizer=custom_tokenizer, min_df=5)
```

```python
X_train_lemma = lemma_vect.fit_transform(text_train)
print('X_train_lemma.shape: {}'.format(X_train_lemma.shape))

vect = CountVectorizer(min_df=5).fit(text_train)
X_train = vect.transform(text_train)
print('X_train.shape: {}'.format(X_train.shape))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

param_grid = {'C': [0.01, 0.1, 1, 10]}
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.99, train_size=0.01, random_state=0)
grid = GridSearchCV(LogisticRegression(), param_grid, cv=cv)

# standard count
grid.fit(X_train, y_train)
print('Best CV score (standard CountVectorizer): {:.3f}'.format(grid.best_score_))

# lemmatization count
grid.fit(X_train_lemma, y_train)
print('Best CV score (lemmatization): {:.3f}'.format(grid.best_score_))
```

```python
vect = CountVectorizer(max_features=10000, max_df=0.15)
X = vect.fit_transform(text_train)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=10, learning_method='batch', max_iter=25, random_state=0)
document_topics = lda.fit_transform(X)
```

```python
print('lda.components_.shape: {}'.format(lda.components_.shape))
lda.components_[0,]
```

```python
sorting = np.argsort(lda.components_, axis=1)[:,::-1]
feature_names = np.array(vect.get_feature_names())
mglearn.tools.print_topics(topics=range(10), feature_names=feature_names, 
                           sorting=sorting, topics_per_chunk=5, n_words=10)
```

```python
topic_1 = np.argsort(document_topics[:,1])[::-1]
for i in topic_1[:10]:
    print(b'.'.join(text_train[i].split(b'.')[:2]) + b'.', '\n')
```

```python
topic_names = ["{:>2} ".format(i) + " ".join(words)
                  for i, words in enumerate(feature_names[sorting[:,:2]])]

fig, ax = plt.subplots(1, 2, figsize=(10,5))
for col in [0,1]:
    start = col*5
    end = (col+1) * 5
    ax[col].barh(np.arange(5), np.sum(document_topics, axis=0)[start:end])
    ax[col].set_yticks(np.arange(5))
    ax[col].set_yticklabels(topic_names[start:end], ha='left', va='top')
    ax[col].invert_yaxis()
    ax[col].set_xlim(0, 10000)
    yax = ax[col].get_yaxis()
    yax.set_tick_params(pad=130)

plt.tight_layout()    
```
